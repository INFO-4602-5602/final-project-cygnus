```python
corr_mat = df.corr()
flat_corr = corr_mat.stack().reset_index()
flat_corr['Topic_Pairs'] = flat_corr.level_0 + ' && ' + flat_corr.level_1
flat_corr.columns = ['A', 'B', 'correlation', 'Topic_Pairs']
flat_corr = flat_corr.loc[flat_corr.correlation < 1, ['Topic_Pairs', 'correlation']]
flat_corr.sort_values(by='correlation',ascending=False, inplace=True)
flat_corr = flat_corr.iloc[::2]
```

```python
import gensim
from gensim import corpora, similarities, models
```

```python
lines = []
with open('reviews.txt') as f:
    for line in f:
        lines.append(line.strip().split())

dictionary = corpora.Dictionary(lines)
    
```

```python
dictionary.save('reviews.dict')
dictionary.filter_extremes(no_below=20, no_above=0.5)
corpus = [dictionary.doc2bow(text) for text in lines]
corpora.MmCorpus.serialize('reviews.mm', corpus)
```

```python
print('Number of unique tokens: %d' % len(dictionary))
print('Number of documents: %d' % len(corpus))
```
